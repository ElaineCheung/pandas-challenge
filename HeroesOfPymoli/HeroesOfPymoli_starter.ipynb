```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players=purchase_data['SN'].nunique()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_unique_items=purchase_data['Item ID'].nunique()
average_price=purchase_data['Price'].mean()
number_purchases=len(purchase_data)
total_revenue=purchase_data['Price'].sum()

average_price="${:,.2f}".format(average_price)
total_revenue="${:,.2f}".format(total_revenue)

purchasing_analysis=pd.DataFrame(
{"Number of Unique Items": number_unique_items,
"Average Price": average_price,
"Number of Purchases": number_purchases,
"Total Revenue":total_revenue},
index=[0])

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data_unique = pd.DataFrame(purchase_data.drop_duplicates(subset=['SN']))
purchase_data_unique
Gender_count = purchase_data_unique['Gender'].value_counts()
Gender_percent= purchase_data_unique['Gender'].value_counts(normalize=True).mul(100).round(2).astype(str)+'%'

gender_demo=pd.DataFrame(
{"Count by Gender": Gender_count,
"Percent by Gender": Gender_percent})

gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group = purchase_data.groupby('Gender')
Purchase_count = gender_group['Purchase ID'].count()
average_purchase_price=gender_group['Price'].mean().round(2)
average_purchase_price=average_purchase_price.astype(str)
average_purchase_price="$"+average_purchase_price
total_purchase_value=gender_group['Price'].sum().round(2)
total_purchase_value_str=total_purchase_value.astype(str)
total_purchase_value_str="$"+total_purchase_value_str
total_players_gender=gender_group['SN'].nunique()
total_purchase_perperson=total_purchase_value/total_players_gender
total_purchase_perperson=total_purchase_perperson.round(2).astype(str)
total_purchase_perperson="$"+total_purchase_perperson
total_purchase_perperson

purchase_gender=pd.DataFrame(
{"Purchase Count": Purchase_count,
"Average Purchase Price": average_purchase_price,
"Total Purchase Value": total_purchase_value_str,
"Avg Total Purchase per Person": total_purchase_perperson})
purchase_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
age_cat = ["<10", "10-14", "15-19", "20-24", "25-29",
           "30-34", "35-39", "40+"]
purchase_data["Age Category"] = pd.cut(purchase_data["Age"], bins, labels=age_cat, include_lowest=True)
purchase_data_unique["Age Category"] = pd.cut(purchase_data["Age"], bins, labels=age_cat, include_lowest=True)
```

```python
# Creating a group based off of the bins
Age_group= purchase_data_unique.groupby("Age Category") 
Age_count=Age_group["Age"].count()
Age_percent=(Age_count/total_players)*100
Age_percent=Age_percent.round(2).astype(str)+'%'
Age_demo=pd.DataFrame(
{"Total Count by Age": Age_count,
"Percentage of Players by Age": Age_percent})
Age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Age_group_all= purchase_data.groupby("Age Category") 

Purchase_count_age = Age_group_all['Purchase ID'].count()
average_purchase_price_age=Age_group_all['Price'].mean().round(2)
average_purchase_price_age=average_purchase_price_age.astype(str)
average_purchase_price_age="$"+average_purchase_price_age
total_purchase_value_age=Age_group_all['Price'].sum().round(2)
total_purchase_value_age_str=total_purchase_value_age.astype(str)
total_purchase_value_age_str="$"+total_purchase_value_age_str
total_players_age= Age_group_all['SN'].nunique()
total_purchase_perperson_age=total_purchase_value_age/total_players_age
total_purchase_perperson_age=total_purchase_perperson_age.round(2).astype(str)
total_purchase_perperson_age="$"+total_purchase_perperson_age
total_purchase_perperson_age

purchase_age=pd.DataFrame(
{"Purchase Count": Purchase_count_age,
"Average Purchase Price": average_purchase_price_age,
"Total Purchase Value": total_purchase_value_age_str,
"Avg Total Purchase per Person": total_purchase_perperson_age})
purchase_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Spender_group = purchase_data.groupby("SN") 
Purchase_count_sn = Spender_group['Purchase ID'].count()
average_purchase_price_sn=Spender_group['Price'].mean().round(2)
total_purchase_value_sn=Spender_group['Price'].sum().round(2)



purchase_sn=pd.DataFrame(
{"Purchase Count": Purchase_count_sn,
"Average Purchase Price": average_purchase_price_sn,
"Total Purchase Value": total_purchase_value_sn})


purchase_sn = purchase_sn.sort_values(by="Total Purchase Value",ascending=False)


purchase_sn['Average Purchase Price'] = purchase_sn['Average Purchase Price'].apply(lambda x: "${:.2f}".format((x)))
purchase_sn['Total Purchase Value'] = purchase_sn['Total Purchase Value'].apply(lambda x: "${:.2f}".format((x)))

top5 = purchase_sn.iloc[0:5]
top5
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
subset=purchase_data[['Item ID','Item Name','Price']]
pop_group =subset.groupby(['Item ID','Item Name'])

Purchase_count_pop = pop_group['Price'].count()
average_purchase_price_pop=pop_group['Price'].mean().round(2)
total_purchase_value_pop=pop_group['Price'].sum().round(2)


purchase_pop=pd.DataFrame(
{"Purchase Count": Purchase_count_pop,
"Average Purchase Price": average_purchase_price_pop,
"Total Purchase Value": total_purchase_value_pop})
purchase_pop

purchase_pop = purchase_pop.sort_values(by="Purchase Count",ascending=False)
purchase_pop


top5_pop = purchase_pop.iloc[0:5]
top5_pop
```
